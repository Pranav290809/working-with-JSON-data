# Working with JSON data

```python
import json
```

```python
# load the data 
def load_data(filename):
    with open(filename , "r") as f:
        data = json.load(f)

    return data
    
```

```python
data = load_data("store_data.json")
# print(data )
# print(type(data))
```

```python
#  data is mixed type 
#  in some points data is missing 
#  in some points data is duplicate
```

```python
# clear and structre the data

def clean_data(data):

    text_to_num = {"one" : 1, "two"  : 2 , "three" : 3 , "four" : 4 , "five" : 5}
    cleaned_data = []
    unique_user = set()
    
    for user in data:
        
        # Clean the rating part 
        raw_rating  = str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            
            raw_rating = text_to_num[raw_rating] 
            
        user["rating"] = raw_rating



        # handle Missing Data
        raw_age = user.get('age')
        if(raw_age == None):
            user['age'] = None


        #working with duplicate data
        if (user["name"].strip() in unique_user ):
            continue
        unique_user.add(user['name'])
        cleaned_data.append(user)

    return cleaned_data
        
```

```python
data = clean_data(data)
# print(data , len(data))
```

```python
def avg_rating(data):
    rating_sum  = 0
    for users in data:
        rating_sum += float(users['rating'])

    return rating_sum / len(data)

        
        
```

```python
rating  = avg_rating(data)
print("avg rating is ",rating)
```

```python
def poor_rating_count(data):
    count = 0
    for users in data:
        if(float(users["rating"]) < 3):
            count+= 1

    return count 
        
```

```python
poor_rating  = poor_rating_count(data)
print(f"percentage of users with poor ratings is {(poor_rating / len(data))* 100 }")
```

```python
# recommandations if rating>= 4 
def user_recom(data):
    recom_dict = {}
    
    for users in data:
        if(float(users["rating"]) >= 4):
            recom_dict[users["name"]] = "recom same brand"
        else:
            recom_dict[users["name"]] = "recom different brand"

    return recom_dict
    
            
```

```python
recom_dict = user_recom(data)
print(recom_dict)
```
